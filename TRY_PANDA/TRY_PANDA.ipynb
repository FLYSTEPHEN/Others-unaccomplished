```python
mapping = {
    '$101-200': 200,
    '$201-400': 400,
    '$401-800': 800,
    '$51-100': 100,
    '$50以下': 50,
}
```

```python
def convert(x):
    return mapping[x]
```

```python
df['price-level']=df['price-range'].apply(convert)
```

```python
df[
    df['country'] == '日本菜'
].sort_values(by = 'price-level',ascending = False)
```
