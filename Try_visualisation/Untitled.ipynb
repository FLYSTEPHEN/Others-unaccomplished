```python
import pandas as pd
df=pd.read_csv('2 jobs-with-frequency.csv')
df = df.fillna(0)
list3=[]
list1=list(df['Description'])
list2=list(df['Location'])
for i in list2:
    list3.append(i.split(', ')[0])
z = list(zip(list3,list1))
```

```python
from pyecharts import Geo

data = z

geo = Geo(
    "gg",
    "data",
    title_color="#fff",
    title_pos="center",
    width=1200,
    height=600,
    background_color="#404a59",
)
attr, value = geo.cast(data)
geo.add(
    "",
    attr,
    value,
    maptype="美国",
    coordinate_region="美国",
    visual_range=[0, 200],
    visual_text_color="#fff",
    symbol_size=15,
    is_visualmap=True,
)
geo.render()
```
